```python
import pandas as pd
```

```python
df = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=["ocorrencia_dia"], dayfirst=True)
df.head()
```

```python
df.replace(['***','****','###!','####','*****'],pd.NA, inplace=True)
```

```python
df
```

```python
#Visualizando dados NA
df.isna().sum()
```

```python
#Filtrando UF NA
df.loc[df.ocorrencia_uf.isna()]
```

```python
#contando os dados
df.count()
```

```python
#Filtrando os dados onde ocorrencia_aerodromo possui dados NA
df.loc[df.ocorrencia_aerodromo.isna()]
```

```python
df.count()
```

```python
#Filtrando linhas onde a ocorrencia_aerodromo possui valores NA e não possui recomendações
filtro1 = df.ocorrencia_aerodromo.isna()
filtro2 = df.total_recomendacoes == 0
df.loc[filtro1 & filtro2]
```

```python
#Filtrando ocorrências do ano 2021 em que houveram incidentes graves
filtro_ano = df.ocorrencia_dia.dt.year == 2021
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
df.loc[filtro1 & filtro_ano]
```

```python
#Filtro buscando ocorrências onde o aerodromo está sem dados e possuem recomendações igual a zero
filtro1 = df.ocorrencia_aerodromo.isna()
filtro2 = df.total_recomendacoes == 0
dfclassificacao = df.loc[filtro1 & filtro2]
```

```python
dfclassificacao.count()
```

```python
#Agrupando as ocorrências onde o aerodromo está sem dados e possuem recomendações igual a zero por classificação
dfclassificacao.groupby(['ocorrencia_classificacao']).codigo_ocorrencia.count()
```

```python
#Agrupando as ocorrências onde o aerodromo está sem dados e possuem recomendações igual a zero por classificação e ordenando da menor quantidade de ocorrências para a maior quantidade de ocorrências

dfclassificacao.groupby(['ocorrencia_classificacao']).size().sort_values()
```

```python
#Filtrando ocorrências do ano de 2021 da região sudeste
filtro_ano = df.ocorrencia_dia.dt.year == 2021
filtro_regiao = df.ocorrencia_uf.isin(['SP', 'MG', 'ES', 'RJ'])
dfsudeste2021 = df.loc[filtro_ano & filtro_regiao]
dfsudeste2021
```

```python
#Filtrando ocorrências do ano de 2021 da região sudeste e agrupando por UF e Classificação

dfsudeste2021.groupby(['ocorrencia_uf', 'ocorrencia_classificacao']).size()
```

```python
#Filtrando ocorrências do ano de 2021 da região sudeste e agrupando por UF, ordenando pela quantidade de ocorrências decrescente

dfsudeste2021.groupby(['ocorrencia_uf']).size().sort_values(ascending=False)
```

```python
#Filtrando ocorrências do ano de 2021 da região sudeste e agrupando por UF e mês
filtro = dfsudeste2021.ocorrencia_classificacao == 'INCIDENTE GRAVE'
dfsudeste2021.loc[filtro].groupby(['ocorrencia_uf', dfsudeste2021.ocorrencia_dia.dt.month]).size()
```
